# Adding a Pandas Dataframe and adding it to SQLITE

```python
# Step 1: Set up the location where you want the sqlite file you make to be. This can be the 
# same folder where the data exists or some other location

sqlite_filename = '/Users/muhammadwaliji/Desktop/personal_projects/polio_data/data_table.sqlite'
```

```python
# Step 2: Create a connection with the SQLite file you created 
# NOTE: SQLITE is essentially a database which means that it can hold various tables inside it

conn = sqlite3.connect(sqlite_filename)
```

```python
# Step 3: Add the DATAFRAME you made above into the SQLite using the .to_sql command
# Make sure that the thing you write before the ".to_sql" is the name of the dataframe
# and makesure to add the connection and if exists to this. Also here is you chance to specify
# what you want to name the table. I called my table "pvac_perc_1yrolds"

#  AT THIS POINT YOU HAVE ADDED THE DATAFRAME YOU MADE ABOVE INTO THE SQLITE FILE
# YOU CAN NOW EITHER CONTINUE TO ADD MORE TABLES OR QUERY THE CURRENT TABLE

# one_yr_olds_long.to_sql("pvac_perc_1yrolds",conn,if_exists="replace",index=False)
```

```python
# Adding one_yr_olds_long to sqlite file
one_yr_olds_long.to_sql("pvac_perc_1yrolds",conn,if_exists="replace",index=False)

# Adding pol_cols to sqlite file
pol_cols.to_sql("polio_instances", conn, if_exists="replace")

# Adding Population data from the country to sqlite file
population_data.to_sql("population",conn,if_exists="replace")

# Adding continent_country data to sqlite file
continent.to_sql("continent_country",conn,if_exists="replace")

# ADDING long_lat to sqlite file
long_lat.to_sql("country_coords", conn, if_exists="replace")
```

```python

continent_population_added=pd.read_sql_query("""
SELECT cc.country_code, cc.name, continent_country.Continent,cc.latitude, cc.longitude, population.population,population.world_share
FROM country_coords AS cc
JOIN continent_country ON continent_country.Country = cc.name
JOIN population ON continent_country.Country = population.country
;
""", conn)

continent_population_added.head()


```

```python
country_coverage_info = pd.read_sql_query("""

SELECT cc.country_code, cc.name, continent_country.Continent,cc.latitude, 
cc.longitude, population.population,population.world_share, pvac_perc_1yrolds.perc_coverage,
pvac_perc_1yrolds.year

FROM country_coords AS cc

JOIN continent_country ON continent_country.Country = cc.name
JOIN population ON continent_country.Country = population.country
JOIN pvac_perc_1yrolds ON pvac_perc_1yrolds.Country =cc.name
;
""", conn)
```

```python
# renaming columns so all start with lower case and dont have any spaces in the name
country_coverage_info.columns=["country_code","name","continent","latitude","longitude","population","perc_pop_world_share","perc_coverage","year"]

#Taking the % sign out of the 'perc_pop_world_share' column so that its easy to convert to int later
country_coverage_info.loc[:,"perc_pop_world_share"]=country_coverage_info["perc_pop_world_share"].str.replace(" %","")

# REPACING NONE IN PERC_COVERAGE COL WITH 0
country_coverage_info.replace(to_replace=[None], value=0, inplace=True)

# Looking at the column head to see if all the changes worked
country_coverage_info.head()
```

```python
# Make a small data frame out of the bigger one to use all the columns I need for analysis
subset = country_coverage_info[["name","continent","year","population","perc_coverage",
                                "perc_pop_world_share"]]

subset.head()
```

```python
# Converting columns with  
subset["perc_pop_world_share"] = pd.to_numeric(subset["perc_pop_world_share"])
subset["year"] = pd.to_numeric(subset["year"])
```
